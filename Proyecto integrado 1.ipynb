##### 5 Prueba de hipotesis
#### Establezco las variables filtradas para mi prueba de Hipótesis.

```python

xone_filtered=games[games['platform']=='XOne']
xone_score=xone_filtered['user_score_fixed']
pc_filtered=games[games['platform']=='PC']
pc_score=pc_filtered['user_score_fixed']

action_filtered=games[games['genre']=='Action']
action_score=action_filtered['user_score_fixed']
sports_filtered=games[games['genre']=='Sports']
sports_score=sports_filtered['user_score_fixed']

```

Verivico si las varianzas son iguales y realizo la prueba de Levene

```python
print(xone_score.var(),pc_score.var())
```

```python
from scipy.stats import levene
statistic, p_value= levene(
    xone_score,pc_score,
   )
p_value
```

No podemos rechazar la hipótesis nula debido a que 0.3229914562269871 es mayor que  el alpha 0.05. En conclucion las varianzas son estadisticamente iguales.

from scipy.stats import ttest_ind
statistic, p_value= ttest_ind(
    sports_filtered['user_score_fixed'],
    action_filtered['user_score_fixed'])
p_value

Realizo la prueba ttest


```python
from scipy.stats import ttest_ind
statistic, p_value= ttest_ind(
   pc_score,xone_score, nan_policy="omit",
   equal_var=True
)
p_value
```

El resultado de la prueba ttest no supera el umbral de 0.05 por lo tanto la hipotesis nula se rechaza.

#### Segunda prueba de hipotesis

```python
from scipy.stats import levene
statistic, p_value= levene(
    action_score,sports_score,
   )
p_value
```

Las varianzas no son iguales en la aprueba de Levene.

```python
from scipy.stats import ttest_ind
statistic, p_value= ttest_ind(
    action_score, 
    sports_score,nan_policy="omit",
    equal_var=False)
p_value
```

 Existe evidencia estadistica o suficiente de que las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes, por lo tanto la hipotesis nula es Rechazada.
